### Dataset Distribution

```python
data_distribution = Skin_noSkin_df.groupby('Y').count()
data_distribution = data_distribution.T.head(1).T
data_distribution
```

`50859` observations of Non skin and `194198` observations with with Skin

```python
data_distribution.plot(kind='bar', legend=False, color=['blue', 'green'])
plt.xlabel('Skin Label')
plt.ylabel('Count')
plt.title('Skin Label Distribution')
plt.xticks(rotation=0)  
plt.show()
```

> Data set contains a lot more Skin observations than non-skin

# Color Channels Distribution 

```python

fig, axes = plt.subplots(3, 1, figsize=(10, 10))

# Red channel histogram
axes[0].hist(Skin_noSkin_df['R'], bins=30, color='red', alpha=0.5)
axes[0].set_title('Distribution of Red')
axes[0].set_xlabel('Color Channel')
axes[0].set_ylabel('Frequency')

# Green channel histogram
axes[1].hist(Skin_noSkin_df['G'], bins=30, color='green', alpha=0.4)
axes[1].set_title('Distribution of Green')
axes[1].set_xlabel('Color Channel')
axes[1].set_ylabel('Frequency')

# Blue channel histogram
axes[2].hist(Skin_noSkin_df['B'], bins=30, color='blue', alpha=0.5)
axes[2].set_title('Distribution of Blue')
axes[2].set_xlabel('Color Channel')
axes[2].set_ylabel('Frequency')





plt.tight_layout()
plt.show()
```

The histograms above shows the pixel frequency distribution for RGB.

Blue Hist: See a peak around 175.

Green Hist: Similar distribution to the Blue and sees a peak at around 175 as well.

Red Hist: The distribution is more spread out with a peak at around 125
